```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read2DPoints(path):
    x, y = [], []
    with open(path, 'r') as file:
        lines = file.read().splitlines()
        for line in lines:
            x1, y1 = line.split()
            x.append(float(x1))
            y.append(float(y1))
    return np.array(x), np.array(y)
```

```python
x, y = read2DPoints('../data/task3-runge-kutta.txt')
plt.title('T3: Runge-Kutta')
plt.plot(x, label='x1')
plt.plot(y, label='x2')
plt.legend()
plt.show()
```

```python
x, y = read2DPoints('../data/task3-runge-kutta-prig.txt')
plt.title('T3: Runge-Kutta prigušeni')
plt.plot(x, label='x1')
plt.plot(y, label='x2')
plt.legend()
plt.show()
```

```python
x, y = read2DPoints('../data/task3-trapezni.txt')
plt.title('T3: Trapezni')
plt.plot(x, label='x1')
plt.plot(y, label='x2')
plt.legend()
plt.show()
```

```python
def normalize(X):
    return (X - min(X)) / (max(X) - min(X))
```

```python
x, y = read2DPoints('../data/task4-runge-kutta-unstable.txt')
plt.title('T4: Runge-Kutta unstable')
plt.plot(normalize(x), label='x1')
plt.plot(normalize(y), label='x2')
plt.legend()
plt.show()
```
